```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get data from CSV file
centers = pd.read_csv("Datasets/Daily_Community_Center_Attendence.csv", index_col="date", parse_dates=True)

# See how many days each center reports attendance
attendance = centers.groupby("center_name").count().drop(columns=['_id'])
print(attendance)

# Create dataframe with only actively reporting centers, threshold of 2000 total reports
above_2000 = attendance[attendance["attendance_count"]>2000]
print(above_2000)

# Plot 10 centers from highest average monthly attendence to lowest
above_2000.plot()
```
